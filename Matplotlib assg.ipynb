```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

```python
x = np.linspace(0, 2, 100) 
y= np.sin(2*math.pi*x)

fig = plt.figure(facecolor=(0.1,0.2,0.1))
plt.subplot(facecolor='#eeffee')
plt.title("Voltage vs. time chart" , color = 'yellow')
plt.xlabel('time(s)', color='white')
plt.ylabel('voltage (mV)', color = 'peachpuff')
plt.plot(x, y)
plt.tick_params(labelcolor='tab:orange')
plt.show()
```

```python
x = np.linspace(0,360,10000) 
y= np.sin(np.deg2rad(x))

plt.figure(facecolor= '#ccddee')


plt.xlim(0,360)
plt.ylim(-1,1)
plt.xticks([0,90,180,270,360])
plt.yticks([-1,-0.5,0,0.5,1])

plt.title("Sine chart")
plt.xlabel('Degree')
plt.ylabel('Sine Value')


plt.grid(True)
plt.grid(True, color='Magenta', ls = '-', lw= 0.5, axis='x')
plt.grid(True, color='Yellow', ls = '-', lw= 0.5, axis='y')
                                                                  
plt.plot(x, y)
```
